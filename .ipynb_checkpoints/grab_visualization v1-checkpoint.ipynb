```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.basemap import Basemap
import numpy as np
%matplotlib inline
```

```python
#import the data
df = pd.read_csv('grab.csv', parse_dates=True)
```

## Filter data for anomalies/outliers

```python
#ignore data with pick-up or drop-off locations outside the philippines
df = df[(df['pick_up_latitude'] <20) & (df['pick_up_latitude'] >5)]
df = df[(df['pick_up_longitude'] <127) & (df['pick_up_longitude'] >117)]
```

# iPhone and Android are the major channels used for booking

```python
print(df.source.value_counts())
df.source.value_counts().plot(kind='bar', figsize=(10,8))
plt.grid(True, alpha=0.2)
plt.title('Grab Booking Source', fontsize=18)
plt.ylabel('Count (Booking Source)', fontsize=14);
```

# Most bookings were done in Metro Manila

```python
print(df.city.value_counts())
df.city.value_counts().plot(kind='bar', figsize=(6,6))
plt.grid(True, alpha=0.2)
plt.title('Cities', fontsize=18)
plt.ylabel('Count (City)', fontsize=14);
```

# Pick-up and Drop-off Locations

```python
plt.figure(figsize=(16,12))
plt.subplot(121)
my_map = Basemap(projection='merc', lat_0=57, lon_0=-135,
    resolution = 'l', area_thresh = 1000.0,
    llcrnrlon=115, llcrnrlat=4,
    urcrnrlon=130, urcrnrlat=21)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='cyan', alpha=0.11)
my_map.drawmapboundary() 
my_map.drawmeridians(np.arange(0, 360, 5), labels=[1,0,0,1])
my_map.drawparallels(np.arange(-90, 90, 5),labels=[1,0,0,1])

lons = list(df.pick_up_longitude)
lats = list(df.pick_up_latitude)
x,y = my_map(lons, lats)
my_map.plot(x, y, 'bo', markersize=5, alpha=0.3)
plt.title('Pick-up Locations', fontsize=22)

#drop_off
plt.subplot(122)
my_map = Basemap(projection='merc', lat_0=57, lon_0=-135,
    resolution = 'l', area_thresh = 1000.0,
    llcrnrlon=115, llcrnrlat=4,
    urcrnrlon=130, urcrnrlat=21)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='cyan', alpha=0.11)
my_map.drawmapboundary() 
my_map.drawmeridians(np.arange(0, 360, 5), labels=[1,0,0,1])
my_map.drawparallels(np.arange(-90, 90, 5),labels=[1,0,0,1])
lons = list(df.drop_off_longitude)
lats = list(df.drop_off_latitude)
x,y = my_map(lons, lats)
my_map.plot(x, y, 'ro', markersize=5, alpha=0.3)
plt.title('Drop-off Locations', fontsize=22);
```

# Booking Time and Date

```python
#Create new Columns for hour and weekday
df['created_at_local'] = pd.to_datetime(df['created_at_local'])
df['day_of_week_num'] = df['created_at_local'].dt.dayofweek
days = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
df['day_of_week'] = df['day_of_week_num'].apply(lambda x:days[x])
#df['month_day_num'] = df['created_at_local'].dt.day
df['hour_of_day'] = df['created_at_local'].dt.hour
```

```python
count_hour_of_day = df.pivot_table(index='hour_of_day',values='city', aggfunc='count')
```

```python
df_comp = df[df['state']=='COMPLETED']
df_una = df[df['state']=='UNALLOCATED']
df_cancel = df[df['state']=='CANCELLED']
comp = df_comp.pivot_table(index='hour_of_day',values='city', aggfunc='count')
una = df_una.pivot_table(index='hour_of_day',values='city', aggfunc='count')
cancel = df_cancel.pivot_table(index='hour_of_day',values='city', aggfunc='count')
df_hr = pd.DataFrame({'index':list(comp.index), 'COMPLETED':list(comp), 'UNATTENDED':list(una), 'CANCELLED':list(cancel)})
df_hr.plot(kind='bar', stacked=True, figsize=(14,8))
df_hr.head()
```

```python
comp = df.pivot_table(index=['hour_of_day', 'state'], values='city', aggfunc='count')
df_comp = comp.to_frame()

df_comp = df_comp.unstack(level='state')
df_comp.plot(kind='bar', stacked=True, figsize=(14,8));
L=plt.legend()
L.get_texts()[0].set_text('CANCELLED')
L.get_texts()[1].set_text('COMPLETED')
L.get_texts()[2].set_text('UNALLOCATED')
plt.ylabel('Total Bookings', fontsize=18)
plt.xlabel('Hour of Day', fontsize=18)
plt.title('Bookings by Hour of Day', fontsize=22)
plt.grid(True, alpha=0.2)
```

```python
count_hour_of_day.plot(kind='bar', figsize=(14,8))
plt.ylabel('Total Journeys', fontsize=18)
plt.xlabel('Hour of Day', fontsize=18)
plt.title('Journeys by Hour of Day', fontsize=22)
plt.grid(True, alpha=0.2)
```

```python
#uber_data['MonthDayNum'] = uber_data['Date/Time'].dt.day
df['month_day_num'] = df['created_at_local'].dt.day

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
df_sept = df[(df['created_at_local']>='2013-09-01') & (df['created_at_local']<='2013-09-30 23:59:59')]
grab_monthdays = df_sept.pivot_table(index=['month_day_num'],
                                  values='city',
                                  aggfunc='count')
grab_monthdays.plot(kind='bar')
plt.ylabel('Count (Bookings)', fontsize=14)
plt.title('Bookings by Month Day (September 2013)', fontsize=18)
plt.grid(True, alpha=0.2)
plt.ylim((0, 15000))

plt.subplot(2,2,2)
df_oct = df[(df['created_at_local']>='2013-10-01') & (df['created_at_local']<='2013-10-31 23:59:59')]
grab_monthdays = df_oct.pivot_table(index=['month_day_num'],
                                  values='city',
                                  aggfunc='count')
grab_monthdays.plot(kind='bar')
plt.title('Bookings by Month Day (October 2013)', fontsize=18)
plt.grid(True, alpha=0.2)
plt.ylim((0, 15000))

plt.subplot(2,2,3)
df_nov = df[(df['created_at_local']>='2013-11-01') & (df['created_at_local']<='2013-11-30 23:59:59')]
grab_monthdays = df_nov.pivot_table(index=['month_day_num'],
                                  values='city',
                                  aggfunc='count')
grab_monthdays.plot(kind='bar')
plt.ylabel('Count (Bookings)', fontsize=14)
plt.title('Bookings by Month Day (November 2013)', fontsize=18)
plt.grid(True, alpha=0.2)
plt.ylim((0, 15000))

plt.subplot(2,2,4)
df_dec = df[(df['created_at_local']>='2013-12-01') & (df['created_at_local']<='2013-12-31 23:59:59')]
grab_monthdays = df_dec.pivot_table(index=['month_day_num'],
                                  values='city',
                                  aggfunc='count')
grab_monthdays.plot(kind='bar')
plt.title('Bookings by Month Day (December 2013)', fontsize=18)
plt.grid(True, alpha=0.2)
plt.ylim((0, 15000))

plt.tight_layout()
```
